```python
from datetime import datetime
messages = []

system_messages = 0
media_messages = 0
deleted_messages = 0

participants = set()

for line in lines:

    line = line.strip()

    if not line:
        continue

    parts = line.split(" - ", 1)

    if len(parts) < 2:
        continue

    timestamp = parts[0]
    rest = parts[1]

    if ": " not in rest:
        system_messages += 1
        continue

    sender, message = rest.split(": ", 1)

    if "<Media omitted>" in message:
        media_messages += 1
        continue

    if "This message was deleted" in message:
        deleted_messages += 1
        continue

    participants.add(sender)

    messages.append({
        "timestamp": timestamp,
        "sender": sender,
        "message": message
    })
start_date = datetime.strptime(messages[0]["timestamp"], "%d/%m/%y, %H:%M")
end_date = datetime.strptime(messages[-1]["timestamp"], "%d/%m/%y, %H:%M")

days = (end_date - start_date).days + 1
print(
    f"Successfully parsed {len(messages)} messages "
    f"from {len(participants)} participants "
    f"over {days} days."
)

print(f"Skipped {system_messages} system messages.")
print(f"Skipped {media_messages} media-omitted messages.")
print(f"Skipped {deleted_messages} deleted messages.")
```

```python
from datetime import datetime
message_data = []

for chat in messages:

    time = datetime.strptime(chat["timestamp"], "%d/%m/%y, %H:%M")

    message_data.append({
        "sender": chat["sender"],
        "time": time,
        "message": chat["message"]
    })
    response_times = {}

for i in range(1, len(message_data)):

    current = message_data[i]
    previous = message_data[i-1]

    if current["sender"] != previous["sender"]:

        gap = (current["time"] - previous["time"]).total_seconds()

        if current["sender"] not in response_times:
            response_times[current["sender"]] = []

        response_times[current["sender"]].append(gap)
        average_response = {}

for person in response_times:

    average = sum(response_times[person]) / len(response_times[person])

    average_response[person] = average
    print("RESPONSE PATTERNS\n")

for person in average_response:

    minutes = average_response[person] / 60

    print(f"{person:<10} : {minutes:.2f} minutes")
    fastest = min(average_response, key=average_response.get)
slowest = max(average_response, key=average_response.get)

print("\nFastest Replier :", fastest)
print("Slowest Replier :", slowest)
```
